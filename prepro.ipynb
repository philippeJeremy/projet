```python
import pandas as pd
```

```python
df = pd.read_json("vin.json")
```

```python
df = df.apply(pd.Series.explode)
df
```

```python
df["vin"] = df["vin"].apply(lambda x : x.replace("\n", ""))
df["vin"] = df["vin"].apply(lambda x : x.replace("\t", ""))
df["vin"] = df["vin"].apply(lambda x : x.replace("\r", ""))
df["region"] = df["region"].apply(lambda x : x.replace("\n", ""))
df["region"] = df["region"].apply(lambda x : x.replace("\t", ""))
df["region"] = df["region"].apply(lambda x : x.replace("\r", ""))
df["region"] = df["region"].apply(lambda x : x.replace(" ", ""))
df["type"] = df["type"].apply(lambda x : x.replace("images/verrevin", ""))
df["type"] = df["type"].apply(lambda x : x.replace(".jpg", ""))

```

```python
df.to_csv("vin.csv")
```
